## Part 2

Sliding window.
Test data:

```text
A: 607 (N/A - no previous sum)
B: 618 (increased)
C: 618 (no change)
D: 617 (decreased)
E: 647 (increased)
F: 716 (increased)
G: 769 (increased)
H: 792 (increased)
```


````python
def count_sliding(measurements: Iterable[int], window: int = 1):
    sliding_sum = [sum(measurements[pos-window:pos])
                   for pos in range(window, len(measurements)+1)]
    # print(sliding_sum)
    return count_increased(sliding_sum)


print(count_sliding((199, 200, 208, 210, 200, 207, 240, 269, 260, 263), 3))
assert count_sliding(
    (199, 200, 208, 210, 200, 207, 240, 269, 260, 263), 3) == 5
````

````python
class Command:
    def __init__(self, input: str) -> None:
        c, x = input.split(' ')
        self.command: str = c.lower()
        self.value: int = int(x)

    def __str__(self) -> str:
        return f'Command {self.command} -> {self.value}'

    def __repr__(self) -> str:
        return self.__str__()


[Command(line) for line in test_input.splitlines() if line]
````

````python
def navigate(commands: Iterable[Command]) -> Tuple[int, int]:
    position = [0, 0]
    for command in commands:
        if command.command == 'forward':
            position[0] += command.value
        elif command.command == 'down':
            position[1] += command.value
        elif command.command == 'up':
            position[1] -= command.value
    return tuple(position)


test_input = """
forward 5
down 5
forward 8
up 3
down 8
forward 2"""

test_commands = [Command(line) for line in test_input.splitlines() if line]
navigate(test_commands)
````

````python
from dataclasses import dataclass


@dataclass
class Position:
    x: int = 0
    y: int = 0

    @property
    def product(self):
        return self.x * self.y
````

````python
def navigate_aim(commands: Iterable[Command]) -> Position:
    position = Position()
    aim = 0
    for command in commands:
        if command.command == 'forward':
            position.x += command.value
            position.y += command.value * aim
        elif command.command == 'down':
            aim += command.value
        elif command.command == 'up':
            aim -= command.value
    return position


test_input = """
forward 5
down 5
forward 8
up 3
down 8
forward 2"""

test_commands = [Command(line) for line in test_input.splitlines() if line]
rv = navigate_aim(test_commands)
print(rv, rv.product)
````

````python
from collections import Counter
def diagnostic(data:Iterable[Iterable[int]]):
    transposed = [[data[j][i] for j in range(len(data))] for i in range(len(data[0]))]
    gamma = bin_iter([Counter(row).most_common()[0][0] for row in transposed])
    epsilon = bin_iter([Counter(row).most_common()[-1][0] for row in transposed])
    return gamma, epsilon

rv = diagnostic(day3_testset)
rv
````

````python
def select_max_or_1(column):
    counter = Counter(column).most_common()
    rv = 1
    if counter[0][1] != counter[1][1]:
        rv = counter[0][0]
    return rv

def select_min_or_0(column):
    counter = Counter(column).most_common()
    rv = 0
    if counter[-1][1] != counter[-2][1]:
        rv = counter[-1][0]
    return rv

print(select_max_or_1([1,1,0]))
print(select_min_or_0([1,1,0]))
print(select_min_or_0([1,0]))
print(select_max_or_1([1,0]))
````

````python
def select_iter(data, column: int, ref_function):
    ref_value = ref_function(row[column] for row in data)
    intermediate_data = [row for row in data if row[column] == ref_value]
    if column < len(data[0]) and len(intermediate_data) > 1:
        intermediate_data = select_iter(intermediate_data, column=column+1, ref_function=ref_function)
    return intermediate_data

test_result = select_iter(day3_testset, column=0, ref_function=select_max_or_1)[0]
print(test_result)
assert test_result == [1,0,1,1,1]
````

````python
def oxy_rating(data: Iterable[Iterable[int]]) -> int:
    result = select_iter(data, column=0, ref_function=select_max_or_1)[0]
    return bin_iter(result), result

oxy_rating(day3_testset)    
````

````python
def scrubber_rating(data: Iterable[Iterable[int]]) -> int:
    result = select_iter(data, column=0, ref_function=select_min_or_0)[0]
    return bin_iter(result), result

scrubber_rating(day3_testset)    
````
